```python
!pip install pandas  scikit-learn
# obs: kaggle para achar base de dados aleatÃ³rias

# Passo 1: Importar a base de dados
import pandas as pd

tabela = pd.read_csv('clientes.csv')
display(tabela)
```

```python
# verificar se tem algum problema com a base de dados
display(tabela.info())
```

```python
# Passo 2: tratamento da base de dados para a IA
from sklearn.preprocessing import LabelEncoder

codificador = LabelEncoder()

# Transformar colunas que sÃ£o texto em tags (label encoder):

# profissao 
#   engenheiro -> 0
#   medico -> 1
#   estudante -> 2...
tabela['profissao'] = codificador.fit_transform(tabela['profissao'])

# mix_credito
tabela['mix_credito'] = codificador.fit_transform(tabela['mix_credito'])

# comportamento_pagamento
tabela['comportamento_pagamento'] = codificador.fit_transform(tabela['comportamento_pagamento'])



```

```python
# x Ã© quem a gente usa para prever
# y Ã© o que a gente quer prever
x = tabela.drop(columns=['score_credito', 'id_cliente'])
y = tabela['score_credito']

from sklearn.model_selection import train_test_split

# DivisÃ£o da base de dados de treino e teste

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3) 
# 70% para treino e 30% para teste; test_size=0.3
# A ordem que as variaviess sÃ£o passadas define seus significados
```

```python
# Passo 3: Criar Modelo de IA: PrevisÃ£o de Score de CrÃ©dito.
# Arvore de decisÃ£o (Random Forest);
# K-Nearest Neighbors (KNN).

# Para usar um IA hÃ¡ 3 passos:

# 1. Importar o modelo de IA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# 2. Criar modelo de IA
modelo_arvore = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

# 3. Treinar modelo de IA.
modelo_arvore.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)




```

```python
# Passo 4: Escolher o melhor modelo
previsao_arvore = modelo_arvore.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

from sklearn.metrics import accuracy_score
display(accuracy_score(y_teste, previsao_arvore))
display(accuracy_score(y_teste, previsao_knn))
```

```python
# Passo 5: Usar o melhor modelo de IA
# 1. Novos clientes
tabela_novos_clientes = pd.read_csv('novos_clientes.csv')

# 2. Ajustar a base de dados
tabela_novos_clientes['profissao'] = codificador.fit_transform(tabela_novos_clientes['profissao'])
tabela_novos_clientes['mix_credito'] = codificador.fit_transform(tabela_novos_clientes['mix_credito'])
tabela_novos_clientes['comportamento_pagamento'] = codificador.fit_transform(tabela_novos_clientes['comportamento_pagamento'])

# 3. Prever o score de crÃ©dito com a Arvore de DecisÃ£o 
# previsao_score_credito = modelo_arvore.predict(tabela_novos_clientes.drop(columns=['id_cliente']))
previsao_score_credito = modelo_arvore.predict(tabela_novos_clientes)
from sklearn.metrics import accuracy_score
display(previsao_score_credito)
```
